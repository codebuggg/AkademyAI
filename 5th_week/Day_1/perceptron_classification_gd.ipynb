```python
# write functions that compute the output of a sigmoid activation function and its derivative
def sigmoid(x):
    fx = 1/(1+np.exp(-x))
    return fx
def sigmoid_d(fx):
    dfx = fx * (1-fx)
    return  dfx
```

```python
def cost_function(a,y):
    loss = -(y * math.log(a) + (1-y) * math.log(1-a))
    return loss
```

```python
# write a training loop

# initialize weights and bias
w1 = np.random.rand()
w2 = np.random.rand()
bias = 0
lrate = 0.01
epochs = 100
m = len(data)

# inside the loop:
for epoch in range(epochs):
    
    for i in range(len(data)):
        point = data[i, :]
    #     print("point: ", point)

    
    # compute output
        z = ((w1 * point[0] )+ (w2 * point[1] )) + bias
        a = sigmoid(z)

      # compute gradients
        dz = a - point[2]
        dw1 += point[0] * dz
        dw2 += point[1] * dz
        db  += dz   
        loss += cost_function(a,point[2])
        
    # compute cost function
    
    
    
    loss = loss/m
    dw1 = dw1/m
    dw2 = dw2/m
    db = db/m

   
    
    # update weights and bias
    w1, w2 = w1 - lrate *dw1, w2 - lrate *dw2
    bias = bias - lrate*db
        
# compute cost
    #print (loss)
# plot costs
#plt.scatter(,loss)
```
